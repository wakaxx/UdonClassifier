```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
import os, time, sys
from ast import literal_eval
driver_path = './driver/chromedriver.exe';
proxies = {
    "http":"",
    "https":""
}
```

```python
search_words = {
    'kake':'かけうどん',
    'kama':'釜揚げうどん',
    'bukkake':'ぶっかけうどん',
    'kayaku':'かやくうどん',
}
```

```python
def get_ulist(search_word,n):
    if n>400:print("n should be less than 400");exit()
    
    driver = webdriver.Chrome(driver_path)
    driver.get("https://www.google.co.jp/imghp?hl=ja&tab=wi&ogbl")
    driver.find_element_by_name("q").send_keys(search_word,Keys.ENTER)
    
    cnt=0
    while (cnt<n):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        page_source= driver.page_source
        soup=BeautifulSoup(page_source,"lxml")
        soup= soup.find_all('div', class_="rg_meta")
        cnt=len(soup)
    else:
        driver.quit()
    
    ulist=[]
    for i in soup:
        dic=i.text.replace("false","False").replace("true","True")
        ulist.append(literal_eval(dic)["ou"])
    
    return ulist[:n]
```

```python
def get_images(folder_nm, urls):
    for cntr,img in enumerate(urls): 
        time.sleep(1)
        print ('[%03d]Downloading.. %s' % (cntr,img))
        try:
            responce = requests.get(img, proxies=proxies)
            with open('%s/%03d.jpg' % (folder_nm, cntr), 'wb') as f:
                f.write(responce.content)
                f.close()
        except:
            pass
    print('Complete!')
    return
```

```python
for folder_nm, search_word in search_words.items():
    FOLDERNAME = './img/' + str(folder_nm)
    if os.path.exists(FOLDERNAME)==False:
        os.mkdir(FOLDERNAME)
    
    urls = get_ulist(search_word, 350)
    time.sleep(10)
    get_images(FOLDERNAME, urls)
```
